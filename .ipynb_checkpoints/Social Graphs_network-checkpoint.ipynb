## Add attributes and parse to a csv

```python
import re
from pathlib import Path
import pandas as pd

# Paths
folder = Path("data/imdb_txt_synopsis_final")
ids = pd.read_csv("selected_movie_ids_1000.csv")
imdb_ids = set(ids["imdb_id"])

def parse_movie_file(path):
    text = path.read_text(encoding="utf-8", errors="ignore")

    def extract(pattern, default=""):
        m = re.search(pattern, text)
        return m.group(1).strip() if m else default

    movieId = extract(r"movieId:\s*(\d+)")
    title   = extract(r"title:\s*(.+)")
    url     = extract(r"url:\s*(.+)")
    imdb_id = re.search(r"tt\d+", url).group(0) if "tt" in url else path.stem
    directors = extract(r"directors:\s*(.+)")
    actors_str = extract(r"actors:\s*(.+)")
    actors = [a.strip() for a in actors_str.split(",")] if actors_str else []

    synopsis = text.split("synopsis:", 1)[-1].strip()

    return {
        "movieId": int(movieId) if movieId else None,
        "imdb_id": imdb_id,
        "title": title,
        "directors": directors,
        "actors": actors,
        "synopsis": synopsis,
    }

rows = []
for path in folder.glob("*.txt"):
    if path.stem in imdb_ids:         # only keep the selected 1000
        rows.append(parse_movie_file(path))

df_movies = pd.DataFrame(rows)
print(df_movies.head())
print(len(df_movies), "movies parsed")
```

## Add genres from Movielens

```python
movies_ml = pd.read_csv("Movielens/movies.csv")   # MovieLens movies file
movies_ml["movieId"] = movies_ml["movieId"].astype(int)

df_movies = df_movies.merge(
    movies_ml[["movieId", "genres"]],
    on="movieId",
    how="left"
)

print(df_movies[["movieId", "title", "genres"]].head())
```

## Add average MovieLens rating (from ratings.csv)

```python
ratings = pd.read_csv("Movielens/ratings.csv")
ratings["movieId"] = ratings["movieId"].astype(int)

avg_ratings = (
    ratings.groupby("movieId")["rating"]
    .mean()
    .reset_index()
    .rename(columns={"rating": "avg_rating"})
)

df_movies = df_movies.merge(avg_ratings, on="movieId", how="left")

print(df_movies[["movieId", "title", "avg_rating"]].head())
```

```python
# ---------- CLEANING BLOCK STARTS HERE ----------

# 1) Clean ACTORS: strip whitespace & drop empty strings
df_movies["actors"] = df_movies["actors"].apply(
    lambda lst: [a.strip() for a in lst if isinstance(a, str) and a.strip() != ""]
)

# 2) Split GENRES into list and main_genre
df_movies["genres_list"] = df_movies["genres"].apply(
    lambda x: x.split("|") if isinstance(x, str) else []
)

df_movies["main_genre"] = df_movies["genres_list"].apply(
    lambda g: g[0] if len(g) > 0 else None
)

# 3) Fix duplicate rating columns (if they exist)
# Sometimes merges create avg_rating_x / avg_rating_y
df_movies = df_movies.drop(columns=["avg_rating_x", "avg_rating_y"], errors="ignore")

# Rename avg_rating → rating for convenience
df_movies = df_movies.rename(columns={"avg_rating": "rating"})

# Quick sanity check
print(df_movies[["movieId", "imdb_id", "title", "main_genre", "rating"]].head())
print("Columns now:", df_movies.columns.tolist())

# ---------- CLEANING BLOCK ENDS HERE ----------
```

## Remove movies with a short synopsis, for a better project

```python
# Compute synopsis length
df_movies["synopsis_length"] = df_movies["synopsis_clean"].str.split().apply(len)

# Filter: keep only movies with >= 20 words
df_movies = df_movies[df_movies["synopsis_length"] >= 100].reset_index(drop=True)

print("Movies remaining after filtering short synopses:", len(df_movies))
```

```python
# LOWERCASE CLEANING BLOCK

# 1. Lowercase the synopsis for sentiment/text analysis
df_movies["synopsis_clean"] = df_movies["synopsis"].str.lower()

# 2. Lowercase genres for consistency
df_movies["genres_list"] = df_movies["genres_list"].apply(
    lambda lst: [g.lower() for g in lst]
)
```

## LABmt sentimental analysis

```python
import pandas as pd

labmt = pd.read_csv(
    "labMT.txt",
    sep="\t",
    comment="#",         # ignore comment-like lines
    header=2,            # header is on line 3 (0-indexed → 2)
    names=[
        "word",
        "happiness_rank",
        "happiness_average",
        "happiness_sd",
        "twitter_rank",
        "google_rank",
        "nyt_rank",
        "lyrics_rank"
    ],
    engine="python"
)

labmt.head()
```

```python
# labmt should already be loaded and have at least: 'word', 'happiness_average'
labmt_dict = dict(zip(labmt["word"], labmt["happiness_average"]))
print("LabMT words in dictionary:", len(labmt_dict))
```

```python
import re
import numpy as np

def labmt_sentiment_from_clean(text):
    """
    Compute average LabMT happiness for a pre-cleaned, lowercase synopsis.
    """
    if not isinstance(text, str) or text.strip() == "":
        return np.nan

    # synopsis_clean is already lowercase, so no .lower() here
    words = re.findall(r"[a-z]+", text)   # only alphabetic words

    scores = [labmt_dict[w] for w in words if w in labmt_dict]

    if not scores:
        return np.nan

    return float(np.mean(scores))
```

## Build bipartite network

```python
# Ensure actors column is list type
df_movies["actors"] = df_movies["actors"].apply(
    lambda x: [a.strip() for a in x] if isinstance(x, list) else []
)
```

```python
df_movies["actors"].head()
```

```python
import networkx as nx

B = nx.Graph()

# Add movie nodes with attributes
for _, row in df_movies.iterrows():
    B.add_node(
        row["imdb_id"],
        bipartite="movie",
        title=row["title"],
        genres=row["genres_list"],
        rating=row["rating"],
        sentiment=row["sentiment_labmt"],
    )

# Add actor nodes + edges
for _, row in df_movies.iterrows():
    movie_id = row["imdb_id"]
    for actor in row["actors"]:
        B.add_node(actor, bipartite="actor")
        B.add_edge(movie_id, actor)
```

```python
print("Nodes:", B.number_of_nodes())
print("Edges:", B.number_of_edges())
```

```python
from networkx.algorithms import bipartite

movies = [n for n, d in B.nodes(data=True) if d["bipartite"] == "movie"]

G = bipartite.weighted_projected_graph(B, movies)
```

```python
print("Movies in network:", G.number_of_nodes())
print("Movie–Movie edges:", G.number_of_edges())
```

```python
for _, row in df_movies.iterrows():
    imdb_id = row["imdb_id"]
    if imdb_id in G:
        G.nodes[imdb_id]["title"] = row["title"]
        G.nodes[imdb_id]["genres"] = row["genres_list"]
        G.nodes[imdb_id]["rating"] = row["rating"]
        G.nodes[imdb_id]["sentiment"] = row["sentiment_labmt"]
```

```python
import community

partition = community.best_partition(G, weight="weight")
nx.set_node_attributes(G, partition, "community")
```

```python
df_movies["community"] = df_movies["imdb_id"].map(partition)
```

```python
from collections import Counter

# Count movies per community
comm_counts = Counter(df_movies["community"])

# Keep only communities with at least 4 movies
valid_communities = {c for c, count in comm_counts.items() if count >= 5}

# Filter dataframe
df_movies = df_movies[df_movies["community"].isin(valid_communities)].copy()

print("Remaining communities:", len(valid_communities))
print(sorted(valid_communities))
```
